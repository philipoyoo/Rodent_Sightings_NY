```python
# Load Data from S3 Bucket
s3 = boto3.resource(
    service_name='s3',
    aws_access_key_id = os.environ.get('s3_access_key'),
    aws_secret_access_key = os.environ.get('s3_secret_access_key')

)
```

```python
for bucket in s3.buckets.all():
    print('Bucket-Name: '+ bucket.name)
```

```python
data = s3.Bucket('fed-practice-data').Object('Rat_Sightings.csv').get()
# Geo_json = s3.Bucket('fed-practice-data').Object('nyc_zip_code.geojson.txt').get()


df = pd.read_csv(data['Body'],index_col=0)
# geo_file = pd.read_csv(Geo_json ['Body'],index_col=0)
df.head()
```

```python

df = df.replace(np.nan, '11206.0', regex=True)
df.head()
```

```python
df['Incident Zip'] = df['Incident Zip'].astype(str).apply(lambda x: x.replace('.0',''))
df=df.reset_index()
df.head()
```
